# Google Drive 저장 노트북
Colab 로컬(`/content/swe3032/output/`)의 결과물을 Google Drive로 복사

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
import os, shutil, glob

SRC_DIR  = '/content/swe3032/output'
DST_DIR  = '/content/drive/MyDrive/swe3032_output'
os.makedirs(DST_DIR, exist_ok=True)
print(f'저장 위치: {DST_DIR}')
```

```python
# 체크포인트(.pth) 복사
pth_files = glob.glob(os.path.join(SRC_DIR, '*.pth'))
for f in pth_files:
    dst = os.path.join(DST_DIR, os.path.basename(f))
    shutil.copy2(f, dst)
    print(f'복사: {os.path.basename(f)}')
print(f'\n체크포인트 {len(pth_files)}개 완료')
```

```python
# 결과 이미지(.png) 복사
png_files = glob.glob(os.path.join(SRC_DIR, '*.png'))
for f in png_files:
    dst = os.path.join(DST_DIR, os.path.basename(f))
    shutil.copy2(f, dst)
    print(f'복사: {os.path.basename(f)}')
print(f'\n이미지 {len(png_files)}개 완료')
```
